```python
# Question 1
```

```python
missing_firstName = df3['firstName'].isnull().sum()
total_firstName = len(df3['firstName'])
percentage_missing_firstName = (missing_firstName / total_firstName) * 100
print("Percentage of missing values in firstName:", percentage_missing_firstName)

missing_lastName = df3['lastName'].isnull().sum()
total_lastName = len(df3['lastName'])
percentage_missing_lastName = (missing_lastName / total_lastName) * 100
print("Percentage of missing values in lastName:", percentage_missing_lastName)


missing_DOB = df3['birthDate'].isnull().sum()
total_DOB = len(df3['birthDate'])
percentage_missing_DOB = (missing_DOB / total_DOB) * 100
print("Percentage of missing values in DOB:", percentage_missing_DOB)
```

```python
# Question 2
```

```python
df3['gender'].fillna(df3['gender'].mode()[0], inplace=True)
female_count = df3[df3['gender'] == 'F'].shape[0]
total_count = df3.shape[0]
female_percentage = (female_count / total_count) * 100
print("Percentage of female gender after imputation:", female_percentage)
```

```python
df4 = df["phoneNumber"]
df4
```

```python
df5 = df4.to_frame()
df5
```

```python
df5 = pd.json_normalize(df5['consultationData'])
df5
```

```python
df_exploded = df5.explode('medicines')
num_medicines = df_exploded['medicines'].count()
num_consultations = df5.shape[0]
average_medicines = num_medicines / num_consultations
print("Average number of medicines prescribed:", average_medicines)
```

```python
# Question 6
```

```python
medicine_counts = df_exploded['medicines'].apply(lambda x: x.get('medicineName') if isinstance(x, dict) else None).value_counts()
third_most_common = medicine_counts.index[2]
print("Third most frequently prescribed medicine:", third_most_common)
```

```python
active_count = df_exploded['medicines'].apply(lambda x: x.get('isActive') if isinstance(x, dict) else None).value_counts().get(True, 0)
inactive_count = df_exploded['medicines'].apply(lambda x: x.get('isActive') if isinstance(x, dict) else None).value_counts().get(False, 0)
total_medicines = active_count + inactive_count
active_percentage = (active_count / total_medicines) * 100
inactive_percentage = (inactive_count / total_medicines) * 100
print("Percentage of active medicines:", active_percentage)
print("Percentage of inactive medicines:", inactive_percentage)
```

```python
def is_valid_indian_number(phone_number):
  if not isinstance(phone_number, str):
    return False
  if len(phone_number) != 10:
    return False
  if not phone_number.isdigit():
    return False
  if not phone_number.startswith(('6', '7', '8', '9')):
    return False
  return True

df3['isValidMobile'] = df3['phoneNumber'].apply(is_valid_indian_number)
print(df3[['phoneNumber', 'isValidMobile']])
```
